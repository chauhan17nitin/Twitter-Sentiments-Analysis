```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv('train_final_matrix_factorized2.csv')
test = pd.read_csv('test_final_matrix_factorized2.csv')
```

```python
y = train['label']
train.drop(['label'], axis = 1, inplace = True)
```

```python
x_train = train.values
y_train = y.values
x_test = test.values
```

```python
import sklearn
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0,1))
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size = 0.2, random_state = 42)
```

```python
import lightgbm as lgb
lgb_train = lgb.Dataset(x_train, y_train)
lgb_eval = lgb.Dataset(x_valid, y_valid, reference = lgb_train)

from sklearn.metrics import f1_score
def lgb_f1_score(y_hat, data):
    y_true = data.get_label()
    y_hat = np.round(y_hat) # scikits f1 doesn't like probabilities
    return 'f1', f1_score(y_true, y_hat), True


evals_result = {}


params = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'num_leaves': 35,
    'max_depth': 20,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 1
}

gbm = lgb.train(params,
                lgb_train,
                num_boost_round=10000,
                valid_sets=lgb_eval,
                feval = lgb_f1_score,
                evals_result=evals_result,
                early_stopping_rounds=150)
lgb.plot_metric(evals_result, metric='f1')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.metrics import f1_score, make_scorer
f1_score = make_scorer(f1_score)
param_grid = {
    'num_leaves': [31, 60, 120],
    'reg_alpha': [0.1, 0.5],
    'min_data_in_leaf': [30, 50, 100, 300],
    'max_depth': [15, 20, 25],
    'learning_rate': [0.3, 0.1, 0.05, 0.01],
    'lambda_l1': [0, 1, 1.5],
    'lambda_l2': [0, 1]
    }
gkf = KFold(n_splits=5, shuffle=True, random_state=42).split(X = x_train, y = y_train)
lgb_estimator = lgb.LGBMClassifier(boosting_type='gbdt',  objective='binary',
                                   num_boost_round=2000, verbose = 1)

gsearch = GridSearchCV(estimator=lgb_estimator, param_grid=param_grid, cv=gkf, scoring = f1_score)
lgb_model = gsearch.fit(X=x_train, y=y_train)
```
