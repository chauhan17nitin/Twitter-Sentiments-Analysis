```python
target = train['label']
train.drop(['label'], axis = 1, inplace = True)
train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0,1))
train = sc.fit_transform(train)
test = sc.transform(test)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(train, target, test_size = 0.2, random_state = 0)
```

```python
from sklearn.model_selection import KFold
import lightgbm as lgb
kf = KFold(n_splits = 3, random_state = 42, shuffle = False)
for train_index, test_index in kf.split(x_train):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = x_train.iloc[train_index], x_train.iloc[test_index]
    Y_train, Y_test = y_train.iloc[train_index], y_train.iloc[test_index]
    
#     print(type(train_index))
#     break
    
    lgb_train = lgb.Dataset(X_train, Y_train)
    lgb_eval = lgb.Dataset(X_test, Y_test, reference = lgb_train)
    
    from sklearn.metrics import f1_score
    def lgb_f1_score(y_hat, data):
        y_true = data.get_label()
        y_hat = np.round(y_hat) # scikits f1 doesn't like probabilities
        return 'f1', f1_score(y_true, y_hat), True
    
    
    evals_result = {}
    
    
    params = {
        'boosting_type': 'gbdt',
        'objective': 'binary',
        'num_leaves': 35,
        'max_depth': 20,
        'learning_rate': 0.1,
        'feature_fraction': 0.9,
        'bagging_fraction': 0.8,
        'bagging_freq': 5,
        'verbose': 0
    }
    
    gbm = lgb.train(params,
                    lgb_train,
                    num_boost_round=10000,
                    valid_sets=lgb_eval,
                    feval = lgb_f1_score,
                    evals_result=evals_result,
                    early_stopping_rounds=150)
    # lgb.plot_metric(evals_result, metric='f1')
    pred_lgb = gbm.predict(X_test)
    x_train_meta = pd.DataFrame(pred_lgb)
    

    import xgboost as xgb
    from sklearn.metrics import f1_score, make_scorer
    dmatrix = xgb.DMatrix(X_train.values,
                         Y_train.values,
                         )
    dtest = xgb.DMatrix(X_test.values,
                         Y_test.values,
                         )
    def f1_eval(y_pred, dtrain):
        y_true = dtrain.get_label()
        score = 1-f1_score(y_true, np.round(y_pred))
        return 'f1_score', score

    evallist  = [(dtest, 'eval')]
    param = {'max_depth':8, 'min_child_weight' : 15, 'eta':0.1,
             'reg_alpha':0.3, 'gamma':0.8, 'colsample_bytree' : 0.5, 'subsample' : 0.9,
             'silent':False, 'objective':'binary:logistic', 'eval_metric' :'error' }
    bst = xgb.train( param, dmatrix, num_boost_round = 2000, early_stopping_rounds=150, evals=evallist, verbose_eval=True, feval = f1_eval)

    d_valid = xgb.DMatrix(X_test.values, )
    pred_xgb = bst.predict(d_valid)
    pred_xgb
    
    x_train_meta['1'] = list(pred_xgb)
    
    x_train_meta['label'] = list(Y_test)
    meta = pd.concat([meta, x_train_meta], ignore_index = True)
    print(meta)
    
#     x_train.iloc[test_index,'lgb'] = pred_lgb
    
#     pd.merge(x_train, x_test_level, left_index = True, right_index = True)
    
    
    
```
