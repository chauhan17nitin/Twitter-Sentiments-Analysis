```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
train.head()
```

```python
test.head()
```

```python
train_y = train['label']
train.drop(['label'], axis = 1, inplace = True)
```

```python
# calculating number of stopwords
import nltk
# nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
train['stopwords'] = train['tweet'].apply(lambda x: len([x for x in x.split() if x in stop]))
test['stopwords'] = test['tweet'].apply(lambda x: len([x for x in x.split() if x in stop]))
test.head()
```

```python
# calculaing number of numerics
train['numerics'] = train['tweet'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
test['numerics'] = test['tweet'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
test.head()
```

```python
# calculating number of uppercases
train['upper'] = train['tweet'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
test['upper'] = test['tweet'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
test.head()
```

```python
train['tweet'] = train['tweet'].apply(lambda x: " ".join(x.lower() for x in x.split()))
test['tweet'] = test['tweet'].apply(lambda x: " ".join(x.lower() for x in x.split()))
test.head()
```

```python
# Removal of stopwords
from nltk.corpus import stopwords
stop = stopwords.words('english')
train['noise1'] = train['tweet'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
test['noise1'] = test['tweet'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
test.head()
```

```python
import re

regex_pattern = "@[\w]*"
def calc_user(cols):
    input_text = str(cols[0])
    urls = re.finditer(regex_pattern, input_text)
    j = 0
    for i in urls: 
        input_text = re.sub(i.group().strip(), '', input_text)
#         print(i.group())
        j = j+1
    return j
def remove_at(cols):
    input_text = str(cols[0])
    urls = re.finditer(regex_pattern, input_text)
    j = 0
    for i in urls: 
        input_text = re.sub(i.group().strip(), '', input_text)
#         print(i.group())
        j = j+1
    return input_text
# remove_regex("remove this @hashtag @from analytics vidhya")
train['user_mention'] = train[['noise1']].apply(calc_user, axis = 1)
test['user_mention'] = test[['noise1']].apply(calc_user, axis = 1)
train['noise1'] = train[['noise1']].apply(remove_at, axis = 1)
test['noise1'] = test[['noise1']].apply(remove_at, axis = 1)
train.head()
```

```python
regex_pattern2 = "#[\w]*"

def compute_hash(cols):
    input_text = str(cols[0])
    
    urls = re.finditer(regex_pattern2, input_text)
    hashtags = [i.group() for i in urls]
    
    hashtags = [j[1:] for j in hashtags]
    return ", ".join(hashtags)
# compute_hash("remove this #hashtag #from analytics vidhya")

def remove_hash(cols):
    input_text = str(cols[0])
    
    urls = re.finditer(regex_pattern2, input_text)
    for i in urls: 
        input_text = re.sub(i.group().strip(), '', input_text)
    return input_text
train['hashtags'] = train[['noise1']].apply(compute_hash, axis = 1)
test['hashtags'] = test[['noise1']].apply(compute_hash, axis = 1)
train['noise1'] = train[['noise1']].apply(remove_hash, axis = 1)
test['noise1'] = test[['noise1']].apply(remove_hash, axis = 1)
test.head()
```

```python
# object Standardisation
lookup_dict = {'rt':'Retweet', 'dm':'direct message', "awsm" : "awesome", "luv" :"love",
               "u" :"you", "ur": "your", "ty": "thankyou" }
def _lookup_words(cols):
    input_text = str(cols[0])
    words = input_text.split() 
    new_words = [] 
    for word in words:
        if word.lower() in lookup_dict:
            word = lookup_dict[word.lower()]
        new_words.append(word) 
    new_text = " ".join(new_words) 
    return new_text
train['noise1'] = train[['noise1']].apply(_lookup_words, axis = 1)
test['noise1'] = test[['noise1']].apply(_lookup_words, axis = 1)
train['hashtags'] = train[['hashtags']].apply(_lookup_words, axis = 1)
test['hashtags'] = test[['hashtags']].apply(_lookup_words, axis = 1)
train.head()
```

```python
import gensim
from gensim.utils import lemmatize

def lemmatize_genism(cols):
    input_text = cols[0]
    output = [wd.decode('utf-8').split('/')[0] for wd in lemmatize(input_text)]
    output_text = " ".join(output)
    return output_text
train['genism_lemm'] = train[['noise1']].apply(lemmatize_genism, axis = 1)
train['hash2'] = train[['hashtags']].apply(lemmatize_genism, axis = 1)
test['genism_lemm'] = test[['noise1']].apply(lemmatize_genism, axis = 1)
test['hash2'] = test[['hashtags']].apply(lemmatize_genism, axis = 1)
train.head()
```

```python
from textblob import TextBlob
train['sentiment'] = train['genism_lemm'].apply(lambda x: TextBlob(x).sentiment[0])
test['sentiment'] = test['genism_lemm'].apply(lambda x: TextBlob(x).sentiment[0])
train.head()
```
